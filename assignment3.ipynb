```python
import numpy as np
import pandas as pd
data = pd.read_csv("IRIS.csv")
print(data)
print(data["species"].unique())
```

```python
def display_stats(Species):
  species_list = data[data["species"]==Species]
  print(f"The Statistics for species {Species} are follows\n")

  print("MEAN:\n")
  mean_sepal_length = species_list["sepal_length"].mean()
  print(f"Mean of sepal length:{mean_sepal_length}\n")
  mean_sepal_width = species_list["sepal_width"].mean()
  print(f"Mean of sepal width:{mean_sepal_width}\n")
  mean_petal_length = species_list["petal_length"].mean()
  print(f"Mean of petal length:{mean_petal_length}\n")
  mean_petal_width = species_list["petal_width"].mean()
  print(f"Mean of petal width:{mean_petal_width}\n\n")

  print("MEDIAN:\n")
  median_sepal_length = species_list["sepal_length"].median()
  print(f"Median of sepal length:{median_sepal_length}\n")
  median_sepal_width = species_list["sepal_width"].median()
  print(f"Median of sepal width:{median_sepal_width}\n")
  median_petal_length = species_list["petal_length"].median()
  print(f"Median of petal length:{median_petal_length}\n")
  median_petal_width = species_list["petal_width"].median()
  print(f"Median of petal width:{median_petal_width}\n\n")

  print("MODE:\n")
  mode_sepal_length = species_list["sepal_length"].mode()
  print(f"Mode of sepal length:{mode_sepal_length}\n")
  mode_sepal_width = species_list["sepal_width"].mode()
  print(f"Mode of sepal width:{mode_sepal_width}\n")
  mode_petal_length = species_list["petal_length"].mode()
  print(f"Mode of petal length:{mode_petal_length}\n")
  mode_petal_width = species_list["petal_width"].mode()
  print(f"Mode of petal width:{mode_petal_width}\n\n")

  print("STANDARD DEVIATION:\n")
  sd_sepal_length = species_list["sepal_length"].std()
  print(f"Standard Deviation of sepal length:{sd_sepal_length}\n")
  sd_sepal_width = species_list["sepal_width"].std()
  print(f"Standard Deviation of sepal width:{sd_sepal_width}\n")
  sd_petal_length = species_list["petal_length"].std()
  print(f"Standard Deviation of petal length:{sd_petal_length}\n")
  sd_petal_width = species_list["petal_width"].std()
  print(f"Standard Deviation of petal width:{sd_petal_width}\n\n")

  print("MINIMUM:\n")
  min_sepal_length = species_list["sepal_length"].min()
  print(f"Minimum of sepal length:{min_sepal_length}\n")
  min_sepal_width = species_list["sepal_width"].min()
  print(f"Minimum of sepal width:{min_sepal_width}\n")
  min_petal_length = species_list["petal_length"].min()
  print(f"Minimum of petal length:{min_petal_length}\n")
  min_petal_width = species_list["petal_width"].min()
  print(f"Mode of petal width:{min_petal_width}\n\n")

  print("MAXIMUM:\n")
  max_sepal_length = species_list["sepal_length"].max()
  print(f"Maximum of sepal length:{max_sepal_length}\n")
  max_sepal_width = species_list["sepal_width"].max()
  print(f"Maximum of sepal width:{max_sepal_width}\n")
  max_petal_length = species_list["petal_length"].max()
  print(f"Maximum of petal length:{max_petal_length}\n")
  max_petal_width = species_list["petal_width"].max()
  print(f"Maximum of petal width:{max_petal_width}\n\n")


display_stats("Iris-setosa")
```
