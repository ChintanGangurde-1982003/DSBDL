```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# data = pd.read_csv("IRIS.csv")
dataset = sns.load_dataset("titanic")
dataset.head()
```

```python
sns.distplot(x=dataset["age"], bins=10,)
```

```python
sns.jointplot(x=dataset["age"],y=dataset["fare"],kind="scatter")
```

```python
sns.jointplot(x=dataset["age"],y=dataset["fare"],kind="hex")
```

```python
sns.barplot(x="sex",y="age",data=dataset, palette=["blue","pink"],estimator=np.std)
```

```python
sns.countplot(x="sex",data=dataset,palette=["blue","pink"])
```

```python
sns.boxplot(x="sex",y="age",data=dataset, palette="Blues",hue="survived")
```

```python
sns.violinplot(x="sex",y="age",data=dataset ,hue="survived")
```

```python
sns.stripplot(x="sex",y="age",data=dataset,jitter=False)
```

```python
sns.stripplot(x="sex",y="age",data=dataset,jitter=True)
```

```python
sns.stripplot(x="sex",y="age",data=dataset,jitter=True,hue="survived")
```

```python
sns.swarmplot(x="sex",y="age",data=dataset,hue="survived")
```

```python
# corr=dataset.corr()
# sns.heatmap(corr,annot=True)
numeric=dataset.select_dtypes(include="number").columns
corr=dataset[numeric].corr()
sns.heatmap(corr,annot=True)
```

```python
sns.histplot(dataset["fare"],kde=True,bins=10)
```
